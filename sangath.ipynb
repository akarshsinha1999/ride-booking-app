Analyze weekly time-series trends for conversion and cancellations to detect 
seasonality. 

```python
import pandas as pd

booking_df["booking_created_at"] = pd.to_datetime(booking_df["booking_created_at"])

booking_df["week"] = booking_df["booking_created_at"].dt.to_period("W").astype(str)

weekly = booking_df.groupby("week").agg(
    total_bookings=("booking_id", "count"),
    completed_count=("status", lambda x: (x == "COMPLETED").sum()),
    cancelled_count=("status", lambda x: (x == "CANCELLED").sum())
).reset_index()

weekly["completed_conversion_rate"] = (
    weekly["completed_count"] * 100 / weekly["total_bookings"]
)

weekly["cancellation_rate"] = (
    weekly["cancelled_count"] * 100 / weekly["total_bookings"]
)

weekly.sort_values("week")
```

The completed conversion rate declines week-over-week (from ~69.2% to ~66.0%), while the cancellation rate increases (from ~30.6% to ~34.0%), indicating a deterioration in booking completion quality.

The second week has significantly higher volume, suggesting that increased demand may be stressing supply, leading to more cancellations.

This pattern points to volume-driven operational strain, not a demand issue, reinforcing the need for capacity and traffic-aware matching during high-demand weeks.

question no 4,

FUNNEL AND STORYTELLING

```python
import numpy as np
import plotly.graph_objects as go


quote_df["rating_segment"] = pd.cut(
    quote_df["driver_rating"],
    bins=[0, 3, 4, 5],
    labels=["Low (≤3)", "Mid (3–4)", "High (4–5)"]
)

stages = ["Quotes", "Bookings", "Completed"]

fig_segmented = go.Figure()

for segment in quote_df["rating_segment"].dropna().unique():
    df_seg = quote_df[quote_df["rating_segment"] == segment]

    q = df_seg["quote_id"].nunique()
    b = df_seg["booking_id"].notna().sum()
    c = (df_seg["booking_status"] == "COMPLETED").sum()

    fig_segmented.add_trace(go.Funnel(
        name=str(segment),
        y=stages,
        x=[q, b, c],
        textinfo="value+percent previous"
    ))

fig_segmented.update_layout(
    title="Booking Funnel Segmented by Driver Rating",
    funnelmode="group"
)

fig_segmented.show()
```
